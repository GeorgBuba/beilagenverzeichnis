```python
# Define a highlight function for the "Titel" column
def highlight_filtered(val):
    """Highlight only rows that do NOT start with the target words."""
    if re.match(pattern, str(val)):
        return ''  # no highlight
    return 'background-color: lightcoral'  # red background for filtered rows
```

```python
# Define a highlight function for empty cells in column "Beilagen_No"
def highlight_empty(val):
    """Highlight empty or NaN cells in orange."""
    if pd.isna(val) or str(val).strip() == '':
        return 'background-color: orange'
    return ''
```

```python
# Apply both styles
styled = (
    df_Beilagen_v1.style
      .applymap(highlight_filtered, subset=['Titel'])
      .applymap(highlight_empty, subset=['Beilagen_No'])
)
```

```python
df_Beilagen_v4 = pd.read_excel('beilagen_Session15_tables_v4.xlsx')
df_Beilagen_v4.head()
```

```python
# Path to your HTML/XML file
file_path = 'https___alex.onb.ac.at_cgi-content_alex_aid=spa&datum=0015&pos=4009&size=45.html'   # change this to your actual file
```

```python
# Read file
with open(file_path, 'r', encoding='utf-8') as f:
    html = f.read()

# Parse with BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Regex pattern: Match exactly "Beilagen " followed by a number range (e.g. 2–10)
pattern = re.compile(r'^Beilagen\s+\d+\s*-\s*\d+$')

data = []
for label in soup.find_all('label', title=True):
    title = label['title'].strip()
    # Keep only titles like "Beilagen 2-10", "Beilagen 383-427", etc.
    if pattern.match(title):
        a_tag = label.find('a', href=True)
        if a_tag:
            href = a_tag['href']
            full_url = 'https://alex.onb.ac.at' + href
            data.append((title, full_url))
```

```python
# Create a DataFrame
df_Links = pd.DataFrame(data, columns=['Title', 'URL'])
df_Links
```

```python
df_Links[['Start', 'End']] = (
    df_Links['Title']
    .str.extract(r'(\d+)\s*-\s*(\d+)')
)

# Coerce to numeric and drop bad rows
df_Links['Start'] = pd.to_numeric(df_Links['Start'], errors='coerce')
df_Links['End'] = pd.to_numeric(df_Links['End'], errors='coerce')

# Drop rows where Start or End could not be converted to a number
df_Links = df_Links.dropna(subset=['Start', 'End']).astype({'Start': int, 'End': int})
```

```python
intervals = pd.IntervalIndex.from_arrays(df_Links['Start'], df_Links['End'], closed='both')

def map_url(num):
    """Map a numeric value to its corresponding URL, or return None."""
    mask = intervals.contains(num)
    if mask.any():
        return df_Links.loc[mask, 'URL'].iloc[0]
    return None
```

```python
# --- Apply mapping only to numeric cells ---
def safe_map(val):
    # Try converting to numeric, skip if fails
    try:
        num = float(val)
        return map_url(num)
    except (ValueError, TypeError):
        return np.nan  # skip text or invalid entries

df_Beilagen_v4['Externer Link'] = df_Beilagen_v4['Beilagen_No'].apply(safe_map)
df_Beilagen_v4.head()
```
